```python
def get_item_id_by_lineage(token, project_id, lineage_urn):
    url = f"https://developer.api.autodesk.com/data/v1/projects/{project_id}/items"
    headers = {"Authorization": f"Bearer {token}"}
    params = {"filter[lineage_urn]": lineage_urn}
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    data = response.json().get("data", [])
    if not data:
        raise Exception(f"File with lineage URN {lineage_urn} not found in project.")
    return data[0]["id"]
```
